```python
import numpy

from datetime import datetime
start=datetime.now()

infinity= float("inf")
global totalCost


global accumCost


global totalCostArr


global stages

startInventory     = 0

S  = 40
Q = 40

shortageCost    = 0

def c(prod,t):
    cost = prod*prodArr[t]
    return cost
    
def h(inventory,t):
    cost = inventory*HoldArr[t]
    return cost


for T in range(1,4):
    totalCost =0
    totalCostArr = []
    accumCost= 0

    stages = T*10
    print()
    print("*********", "T=", stages, "**********")
    f = numpy.zeros([stages + 2, S + 1])
    x = numpy.zeros([stages + 1, S + 1], dtype=int)
    
    HoldArr= numpy.random.uniform(low=5.0,high=15.0,size=stages+1)
    print("holding cost array for stages:", HoldArr)

    print()

    prodArr= numpy.random.uniform(low=10.0,high=20.0,size=stages+1)
    print('ordering cost array for stages:', prodArr)
    
    d=numpy.random.randint(low=0,high=20,size=stages+1,dtype=int)
    print()
    print('demand array:',numpy.random.randint(low=0,high=20,size=stages+1,dtype=int))

    for data in range(1,11):
        print('iteration',data)


        for t in range(stages,0,-1) :

                for i in range(S + 1) :

                    minxt = max(0, d[t] - i) #the minxt = the max of 0 and d(t) - i becuase order cannot be negative.
                    maxxt = min(Q, S - i + d[t]) #else we exceed S


                    optimalValue = infinity #we start by setting an OptimalValue = infinity so that in each iteration, we can select the lower cost.

                    for theRange in range(minxt, maxxt+1) : 
                        
                        if(i + theRange - d[t] <0):
                            print("next inv",d[t]-i-theRange)
                            
                        else:    
                            j = i + theRange - d[t] #j = the next state. So next state is j if current inventory = i and the dt is shipped off.

                            #if (i<dt(t+1)) :
                               # cost += shortageCost

                            costOption = h(i-d[t]+theRange,t) + c(theRange,t) + f[t+1,j]

                            if costOption < optimalValue :
                                optimalValue = costOption
                                optimalMove = theRange
                            f[t,i] = optimalValue
                            x[t,i] = optimalMove
                

        for t in range(stages,0,-1):
            print ("*****","week",t,"demand",d[t],"*****")
            print()
            for i in range(S+1):
                    print("begInv", i)
                    print('remain Inv',i + x[t,i] - d[t])
                    print("c:",f[t,i])
                    print("opt order:",x[t,i])
                    print()
                    
                    solutionString = "when we order the following quantities in each week t:"
                    i = startInventory

        for t in range(1,stages+1) : 
            solutionString += " " + str(x[t,i])
            i = i + x[t,i] - d[t] #ending inventory 
            print("The cost for week",t," is: " + str(f[t,startInventory]))
            totalCost = totalCost + f[t,startInventory]

        print("minimized cost is:", totalCost)

        print(solutionString)

        accumCost= accumCost +totalCost
        totalCostArr.append(totalCost)

    print("accumCost",accumCost)
    print("average", accumCost/10)
    q=numpy.std(totalCostArr)
    print("std:",q)

print( datetime.now()-start )
```
